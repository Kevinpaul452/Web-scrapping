
# **Data Acquisition from selected website.**


The most important aspects of the webpage were then extracted as follows:

* URL of the website 
* Inspected the html page 
* Selected the important class

```python
resp  #Check HTTP Status
```

```python
resp.content
```

```python
soup = BeautifulSoup(resp.content, 'html.parser')           #Creating a Soup object for scraping
```

```python
ul = soup.find_all('div',{'class':'posts grid'})       #storing the first page's target results
```

```python
ul
```

```python
soup
```

```python
results = soup.find_all('div', attrs={'class':'sr-grid-cell quick-peek-container'})
```

```python
len(results)    #Checking the length of results
```

```python
results[0]      #Checking the first
```

```python
get_titles=soup.findAll('div',{'class':'title'})
```

```python
len(get_titles)
```

```python
titles=[]
for item in get_titles:
    titles.append(item.text)
titles=pd.Series(titles)
```

```python
get_location=soup.findAll('div',{'class':'location'})
```

```python
len(get_location)
```

```python
get_location[22].text.replace('\n','')           #Finding the location name 
```

```python
location=[]
for item in get_location:
    location.append(item.text)
location=pd.Series(location)
```

```python
get_prices=soup.findAll('div',{'class':'price'})
```

```python
prices=[]
for item in get_prices:
    prices.append(item.text)
prices=pd.Series(prices)
```

```python
data=pd.concat([prices,titles,location],axis=1)
```

```python
data.columns=['Prices','Titles','Location']
```

```python
data
```

# **Data Preprocessing**

In the process of data mining, one of the most significant steps is called preprocessing, and it involves the altering or dropping of data prior to its usage in order to assure or increase performance.



```python
data.at[0,'Prices'].replace('\n','').replace('€','')
```

```python
data['Prices'] = data['Prices'].str.replace(r"[\€]", '')
```

```python
data.head()    #Showing the head of the data
```

```python
data['Titles'] = data['Titles'].str.replace(r"\n", '')
```

```python
data.head()
```

```python
data['Location'] = data['Location'].str.replace(r"\n", '')
```

```python
data
```

```python
data.describe()
```

```python
data.shape  #Shows the shpe of data
```

```python
data.dtypes #Need to replace the Prices column from object to float
```

```python
data['Prices']=data['Prices'].replace('[\n,]','', regex=True).astype(float) #Removing unwanted strings and replacing Prices column from object to float
```

```python
data.dtypes
```

```python
data = data.drop([30,31],axis=0) #Droping the rows 30 and 31
```

```python
data[data['Prices']>200] #Checking the Prices values greater than 200
```
